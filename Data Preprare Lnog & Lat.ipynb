```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
district = "Nadia"
state = "West Bengal"
country = "India"
```

```python
# Get coordinates of district -- MAPMYINDIA service
def getCoordinates(district, state, country):
    url = f"https://atlas.mapmyindia.com/api/places/geocode?address={district},{state},{country}&itemCount=1&bias=0&podFilter=dist"
    r = requests.get(url,headers={"accept":"application/json","Authorization": "Bearer XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"})
    return json.loads(r.text)["copResults"]["longitude"], json.loads(r.text)["copResults"]["latitude"]
```

```python
getCoordinates(district,state,country)
```

```python
df = pd.read_excel("districts-db-form.xlsx")
```

```python
df.head()
```

```python
for i in range(148,df.shape[0]):
    district = df.iloc[i]["district_name"]
    state = df.iloc[i]["state_name"]
    country = "India"
    
    longitude , latitude = getCoordinates(district,state,country)
    print(f"{i} {longitude} {latitude}")
    df.at[i,"latitude"] = latitude
    df.at[i,"longitude"] = longitude    
```
